All 23 uniques values of three columns **segment_tag1**, **segment_tag2** and **spending_category2** are predicted user segments from predicive model. Users with the same predictive persona tend to have the same transaction behaviour.

```python
# Function to combine segment tags
def combine_tags(row):
  tags = set()
  for col in ['segment_tag1', 'segment_tag2', 'segment_tag3']:
    tags.add(str(row[col]))
  return ','.join(tags)

# Apply the function to create the new 'segment_tag' column
data['segment_tag'] = data.apply(combine_tags, axis=1)

# Drop the original segment tag columns
data.drop(columns=['segment_tag1', 'segment_tag2', 'segment_tag3'], inplace=True)

# Display the first few rows to verify the changes
data.head()
```

```python
# Check if any segment_tag has more than 3 elements
def check_segment_length(segment):
    return len(segment.split(',')) > 3

data['segment_tag_length'] = data['segment_tag'].apply(check_segment_length)
print(data['segment_tag_length'].value_counts())

if data['segment_tag_length'].any():
    print("At least one segment_tag has more than 3 elements.")
else:
    print("No segment_tag has more than 3 elements.")

data.drop(columns=['segment_tag_length'], inplace=True)
```

This seems Ok.

```python
def check_user_segments(data):
    """
    For each user ID, collects corresponding segment tags in a set and checks if any set has more than 3 elements.
    """
    user_segments = {}
    for index, row in data.iterrows():
        user_id = row['user_id']
        segment_tag = row['segment_tag']
        if user_id not in user_segments:
            user_segments[user_id] = set()
        user_segments[user_id].add(segment_tag)

    for user_id, segments in user_segments.items():
      if len(segments) > 3:
        print(f"User {user_id} has more than 3 segments: {segments}")
        return
    print("No user has more than 3 segments.")

check_user_segments(data)
```
